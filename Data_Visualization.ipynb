## Line Plot

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

### plotting the *graph*

```python
# plotting a simple function
```

```python
# using list
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]
```

```python
plt.plot(year,price)
```

```python
# plotting from a pandas dataframe

batsman = pd.read_csv('/content/sharma-kohli.csv')
batsman.head(3)
```

```python
## single graph plot
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
## plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

### Lables Title

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

#lables

plt.title('Rohit Sharma Vs V.Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

### Color (hex) , Line (width & style) and Marker (size)

```python
# color
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green')
plt.plot(batsman['index'],batsman['RG Sharma'], color = 'blue')
```

```python
# line (width & style)

plt.plot(batsman['index'],batsman['V Kohli'],linestyle = 'dashdot', linewidth = 2 )
plt.plot(batsman['index'],batsman['RG Sharma'], linestyle= 'solid', linewidth=1)
```

```python
# marker(size)

plt.plot(batsman['index'],batsman['V Kohli'], marker = '*', markersize=5)
plt.plot(batsman['index'],batsman['RG Sharma'], marker = '*',markersize=9)
```

###  legend -> location

```python
# define legend
plt.plot(batsman['index'],batsman['V Kohli'],label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],label='Rohit')

# show legend

plt.legend(loc='upper center')
```

### limiting axes


```python
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)

price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)# outlier make graph meaningless
```

```python
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)

#limit y axis

plt.ylim(40000,70000)
plt.xlim(2016,2020)
```

### Grid

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

# grid
plt.grid()
```

### show graph

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.grid()

plt.show()
```

## Scatter Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### plotting scatter plot

```python
x = np.linspace(-10,10,50)
y=10*x + 3 + np.random.randint(0,300,50)
```

```python
# 1.
plt.scatter(x,y)
```

```python
# 2. Using pandas data

df = pd.read_csv('/content/batter.csv')
df.head()
```

```python
plt.scatter(df['avg'],df['strike_rate'])

# label
plt.title('Avg & Strike Rate of batsman')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')
```

### Scatter parameters like s,marker etc.

```python
# s
import seaborn as sns
tips = sns.load_dataset('tips')
tips.sample(4)
```

```python
# size
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*15)
```

```python
# color & marker

plt.scatter(tips['total_bill'],tips['tip'],color='#9ACD32', marker='*')
```

### Scatter using plt.plot

```python
plt.plot(tips['total_bill'],tips['tip'],'o')
```

**plt.plot Vs plt.scatter**



*   plt.plot faster than plt.scatter
*   plt.scatter have lots parameter like color, marker, s etc. that not support in plt.plot



## Bar Chart

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

### plotting bar chart

```python
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children)
```

```python
# Horizontal bar chart

plt.barh(colors,children)
```

### color and label

```python
df = pd.read_csv('/content/batsman_season_record.csv')
df.head(3)
```

```python
plt.bar(df['batsman'],df['2015'])
```

### multiple bar chart

```python
plt.bar(np.arange(len(df.batsman))-0.3,df['2015'],width=0.3)
plt.bar(np.arange(len(df.batsman)),df['2016'],width=0.3)
plt.bar(np.arange(len(df.batsman))+0.3,df['2017'],width=0.3)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

### xticks

```python
children = [10,20,40,10,30]
colors = ['red_red','blue_blue','green_green','yellow_yellow','pink_pink']

plt.bar(colors,children)
```

```python
children = [10,20,40,10,30]
colors = ['red_red','blue_blue','green_green','yellow_yellow','pink_pink']

plt.bar(colors,children)

#xticks
plt.xticks(rotation='vertical')

plt.show()
```

### Stacked bar chart

```python
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom = df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom = (df['2016']+df['2017']),label='2015')

plt.legend()
plt.show()
```

## Pie Chart

### Create pie chart

```python
data = [23,45,100,20,49]

plt.pie(data)

plt.show()
```

```python
# labels
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset

df = pd.read_csv('/content/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'])
plt.show()
```

### Percentage(Autopect)

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

### Color

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

### explode shadow

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],explode=[0.3,0,0,0,0,0.1])

plt.show()
```

```python
# shadow

plt.pie(df['batsman_runs'],labels=df['batsman'],explode=[0.3,0,0,0,0,0.1],shadow=True)

plt.show()
```
